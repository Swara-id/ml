```python
import cv2
import mediapipe as mp
import pandas as pd
import os
```

# For Experiment purpose 

```python
# Define drawing and landmark utility functions
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# Initialize hand landmark detection with specific parameters
hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5, max_num_hands=2)

# Video capture setup
cap = cv2.VideoCapture(0)

while cap.isOpened():
    success, image = cap.read()
    if not success:
        print("Ignoring empty camera frame.")
        continue

    # RGB to BGR conversion for MediaPipe
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Hand landmark detection
    results = hands.process(image)

    # Draw hand landmarks and bounding box on the image
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            # Get hand bounding box coordinates
            mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
            print(hand_landmarks)

    # Display the processed image with landmarks
    cv2.imshow('Test', cv2.flip(image, 1))
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break
 
# When all the process is done
# Release the capture and destroy all windows
capture.release()
cv2.destroyAllWindows()
```

# Code to Extracting hand landmarks Coordinate from each Images

```python
# Define drawing and landmark utility functions
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# Initialize hand landmark detection with specific parameters
hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5, max_num_hands=2)

# Directory containing the datasets
data_dir = 'datasets/Filtered Raw Images'

# Initialize an empty list to store the landmarks
landmark_list = []

# Process each subfolder (label) in the dataset directory
for label in os.listdir(data_dir):
    label_path = os.path.join(data_dir, label)
    if os.path.isdir(label_path):
        # Process each image in the subfolder
        for image_file in os.listdir(label_path):
            image_path = os.path.join(label_path, image_file)
            image = cv2.imread(image_path)
            if image is None:
                continue
            
            # RGB to BGR conversion for MediaPipe
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            # Hand landmark detection
            results = hands.process(image_rgb)

            # Extract and store hand landmarks if detected
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    landmark_data = {'label': label}
                    for idx, landmark in enumerate(hand_landmarks.landmark):
                        landmark_data[idx] = [landmark.x, landmark.y, landmark.z]
                    
                    # Append to the list
                    landmark_list.append(landmark_data)

# Convert the list to a DataFrame
landmark_df = pd.DataFrame(landmark_list)

# Save the DataFrame to a CSV file (optional)
landmark_df.to_csv('./datasets/hand_landmarks.csv', index=False)

# Release resources
hands.close()

print("Processing complete. Data saved to hand_landmarks.csv")
```

```python
import cv2
import mediapipe as mp
import pandas as pd
import os

# Define drawing and landmark utility functions
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

# Initialize hand landmark detection with specific parameters
hands = mp_hands.Hands(min_detection_confidence=0.5, min_tracking_confidence=0.5, max_num_hands=2)

# Directory containing the datasets
data_dir = 'datasets/Filtered Raw Images'

# Finger names mapping for MediaPipe hand landmarks
finger_names = {
    0: "Wrist",
    1: "Thumb",
    2: "Thumb",
    3: "Thumb",
    4: "Thumb",
    5: "Index",
    6: "Index",
    7: "Index",
    8: "Index",
    9: "Middle",
    10: "Middle",
    11: "Middle",
    12: "Middle",
    13: "Ring",
    14: "Ring",
    15: "Ring",
    16: "Ring",
    17: "Pinky",
    18: "Pinky",
    19: "Pinky",
    20: "Pinky"
}

# Initialize an empty list to store the landmarks
landmark_list = []

# Process each subfolder (label) in the dataset directory
for label in os.listdir(data_dir):
    label_path = os.path.join(data_dir, label)
    if os.path.isdir(label_path):
        # Process each image in the subfolder
        for image_file in os.listdir(label_path):
            image_path = os.path.join(label_path, image_file)
            image = cv2.imread(image_path)
            if image is None:
                continue
            
            # RGB to BGR conversion for MediaPipe
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            # Hand landmark detection
            results = hands.process(image_rgb)

            # Extract and store hand landmarks if detected
            if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    for idx, landmark in enumerate(hand_landmarks.landmark):
                        landmark_data = {
                            'label': label,
                            'X': landmark.x,
                            'Y': landmark.y,
                            'Z': landmark.z,
                            'finger_name': finger_names[idx],
                            'landmark_index': idx
                        }
                        # Append to the list
                        landmark_list.append(landmark_data)

# Convert the list to a DataFrame
landmark_df = pd.DataFrame(landmark_list)

# Save the DataFrame to a CSV file (optional)
landmark_df.to_csv('hand_landmarks perfected.csv', index=False)

# Release resources
hands.close()

print("Processing complete. Data saved to hand_landmarks.csv")
```
